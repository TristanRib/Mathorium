```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go

from src.classes.EscherichiaColy import EscherichiaColy, FastFood
```

```python
bacteria = EscherichiaColy(0, 0)
targets = [FastFood(30, 30), FastFood(-30, 30), FastFood(30, -30), FastFood(-30, -30)]
```

```python
path = bacteria.chemotaxis(targets, 1000)
```

```python
x, y = zip(*path)

plt.figure(figsize=(8, 6))

plt.plot(x, y, marker='o', linestyle='-', color='blue', label="Trajectoire")
plt.scatter(x[0], y[0], color='green', s=100, label="Départ", zorder=5)
plt.scatter(x[-1], y[-1], color='orange', s=100, label="Arrivée", zorder=5)

cible_label_added = False
for target in targets:
    tx, ty = target.position
    if not cible_label_added:
        plt.scatter(tx, ty, color='red', marker='x', s=100, label="Cibles", zorder=4)
        cible_label_added = True
    else:
        plt.scatter(tx, ty, color='red', marker='x', s=100, zorder=4)

    cercle = plt.Circle((tx, ty), target.radius(), color='red', alpha=0.2)
    plt.gca().add_patch(cercle)

# Mise en forme
plt.title("Trajectoire 2D de la bactérie")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```
